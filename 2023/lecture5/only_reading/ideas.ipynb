Вот два примера кода, демонстрирующих ООП в C++:

**Пример 1: Класс "Студент"**
```cpp
#include <iostream>
#include <string>

class Student {
private:
    std::string name;
    std::string surname;
    int age;
    float averageGrade;

public:
    Student(const std::string& name, const std::string& surname, int age, float averageGrade)
        : name(name), surname(surname), age(age), averageGrade(averageGrade) {}

    void PrintInfo() {
        std::cout << "Name: " << name << std::endl;
        std::cout << "Surname: " << surname << std::endl;
        std::cout << "Age: " << age << std::endl;
        std::cout << "Average Grade: " << averageGrade << std::endl;
    }
};

int main() {
    Student student("John", "Doe", 20, 85.5);
    student.PrintInfo();

    return 0;
}
```

**Пример 2: Класс "Круг"**

```cpp
#include <iostream>

class Circle {
private:
    double radius;

public:
    Circle(double radius) : radius(radius) {}

    double GetRadius() const {
        return radius;
    }

    void SetRadius(double newRadius) {
        radius = newRadius;
    }

    double CalculateArea() const {
        return 3.14159 * radius * radius;
    }

    double CalculateCircumference() const {
        return 2 * 3.14159 * radius;
    }
};

int main() {
    Circle circle(5.0);
    std::cout << "Radius: " << circle.GetRadius() << std::endl;
    std::cout << "Area: " << circle.CalculateArea() << std::endl;
    std::cout << "Circumference: " << circle.CalculateCircumference() << std::endl;

    return 0;
}
```
